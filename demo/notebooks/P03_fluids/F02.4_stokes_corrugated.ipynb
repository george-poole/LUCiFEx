# Flow of a Stokes fluid in a corrugated channel

$$
\mathbb{S}
\begin{cases}
\Omega = \left\{(x, y)~:~0<x<L_x~,~\epsilon\sin\left(\frac{2\pi nx}{L_x}\right)<y<L_y \right\} \\
\partial\Omega_{\text{lower}} = \left\{(x, y)~:~ y=\epsilon\sin\left(\frac{2\pi nx}{L_x}\right) \right\} \\
\partial\Omega_{\text{upper}} = \{(x, y)~:~y=L_y \} \\
\partial\Omega_{\text{left}} = \{(x, y)~:~x=0 \} \\
\partial\Omega_{\text{right}} = \{(x, y)~:~x=L_x \} \\
\textbf{u}_{\text{E}}\vert_{\partial\Omega_{\text{lower}}\cup\partial\Omega_{\text{upper}}}=\textbf{0} & \text{no-flow on upper and lower boundaries} \\
\boldsymbol{\tau}_{\text{N}}(x=0,y)=p_{\text{in}}\textbf{e}_x & \text{equivalent to } p(x=0,y)=p_{\text{in}}\\ 
\boldsymbol{\tau}_{\text{N}}(x=L_x,y)=\textbf{0} & \text{equivalent to } p(x=L_x,y)=0\\ 
\end{cases}
$$

```python
import numpy as np
from lucifex.mesh import mesh_from_boundaries
from lucifex.viz import plot_mesh, plot_line

h = 0.05
Lx = 4.0
Ly = 1.0
eps_lower = 0.1
eps_upper = 0.1
n_lower = 10
n_upper = -4

lower = lambda x: eps_lower * np.sin(np.pi * n_lower * x / Lx)
upper = lambda x: eps_upper * np.sin(np.pi * n_upper * x / Lx)
boundaries = (
    lambda x: x[0] - Lx,
    lambda x: x[1] - Ly - upper(x[0]),
    lambda x: x[0],
    lambda x: x[1] - lower(x[0]),
)
dx_bbox = 1e-2
dy_bbox = 1e-2
x_bbox = np.arange(0, Lx + dx_bbox, dx_bbox)
y_bbox = np.arange(-eps_lower, Ly + eps_upper + dy_bbox, dy_bbox)

mesh = mesh_from_boundaries(h, 'triangle')(x_bbox, y_bbox, *boundaries)
fig, ax = plot_mesh(mesh)
plot_line(
    fig, ax, 
    [(x_bbox, lower(x_bbox)), (x_bbox, Ly + upper(x_bbox))], 
    color='blue',
)
```
