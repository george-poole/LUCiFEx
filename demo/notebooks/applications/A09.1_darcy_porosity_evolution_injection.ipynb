# Porosity evolution-dissolution from an injected Darcy fluid in a rectangle

$$
\mathbb{S}
\begin{cases}
\Omega = \left[-\tfrac{1}{2}L_x, \tfrac{1}{2}L_x\right] \times [0, L_y] \\
c_0 = 1 \\
0\leq b_0 \leq 1 \\
\phi_0 = 1 \\
u_{\text{E}}(x,y=0) =
\begin{cases}
-1 & |x| \leq \tfrac{1}{2}\delta L_x \\
0 & \text{otherwise}
\end{cases}  & \text{inflow velocity boundary condition} \\
u_{\text{E}}(x,y=L_y) = \delta & \text{outflow velocity boundary condition} \\
u_{\text{E}}(x=0,y) = u_{\text{E}}(x=L_x,y) = 0 & \text{no-penetration boundary condition} \\
c(|x| \leq \tfrac{1}{2}\delta L_x, y=0) = 0 & \text{inflow concentration boundary condition} \\
b(|x| \leq \tfrac{1}{2}\delta L_x, y=0) = b_{\text{in}} \\ 
\frac{\partial c}{\partial x}\bigg\vert_{|x| > \tfrac{1}{2}\delta L_x,y=0}=\frac{\partial c}{\partial x}\bigg\vert_{x,y=L_y}=\frac{\partial c}{\partial x}\bigg\vert_{x=\pm\tfrac{1}{2}L_x,y}=0 & \text{no-flux on other boundaries}\\
\frac{\partial b}{\partial x}\bigg\vert_{|x| > \tfrac{1}{2}\delta L_x,y=0}=\frac{\partial b}{\partial x}\bigg\vert_{x,y=L_y}=\frac{\partial b}{\partial x}\bigg\vert_{x=\pm\tfrac{1}{2}L_x,y}=0 \\
\mathsf{K} = \phi^2\mathsf{I} \\ 
\mathsf{D}_b = \mathsf{I} \\ 
\mathsf{D}_c = \mathsf{I} \\ 
\mu = 1 \\
\rho(b, c) = c + \beta b 
\end{cases}
$$
