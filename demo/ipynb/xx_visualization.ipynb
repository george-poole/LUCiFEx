### Visualization

```python
from ufl.core.expr import Expr
from dolfinx.fem import FunctionSpace
from lucifex.mesh import rectangle_mesh
from lucifex.fem import StaticFunction
from lucifex.viz import plot_colormap, plot_line, plot_cross_section

mesh = rectangle_mesh(2.0, 1.0, 10, 10)
fs = FunctionSpace(mesh, ('P', 1))
```

$f(x, y) = x^2 - y^2 + xy$

```python
f = StaticFunction(fs, lambda x: x[0]**2 - x[1]**2 + x[0] * x[1], name='f')
fig, ax = plot_colormap(f, 'f(x,y)')
```

```python
fig, ax = plot_cross_section(f, 'x', fraction=0.5)
fig, ax = plot_cross_section(f, 'y', fraction=0.8)
```
