```python
import numpy as np
from IPython.display import HTML

from lucifex.viz import create_animation, plot_line
from lucifex.io import write


x_axis = np.linspace(0, 1, num=200)
times = np.linspace(0, 2, num=200)
xfunc = lambda x, t: np.sin(2 * np.pi * (x - t))

line_series = [
    (x_axis, xfunc(x_axis, t)) for t in times 
]
title_series = [f'$t={t:.2f}$' for t in times]

anim = create_animation(
    plot_line,
    x_label='$x$',
    y_label='$u(x,t)$',
)(line_series, title=title_series)
```

```python
HTML(anim.to_html5_video(), './figures')
```
