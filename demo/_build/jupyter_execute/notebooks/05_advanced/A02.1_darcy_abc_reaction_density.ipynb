# ABC convection with a destabilising reaction

$$
\mathbb{S}
\begin{cases}
\mathcal{L} = \sqrt{D_{\text{ref}}\mathcal{T}/\phi_{\text{ref}}} & \text{scaling with diffusive length and reactive time} \\
\mathcal{U} = \phi_{\text{ref}}\mathcal{L}/\mathcal{T} \\
\mathcal{T} = \phi_{\text{ref}}\Delta c/\Delta R \\
\Omega = [0, \sqrt{RaDa}\mathcal{A}] \times [0, \sqrt{RaDa}] & \text{aspect ratio } \mathcal{A}=\mathcal{O}(1) \\
\textbf{e}_g=-\textbf{e}_y & \text{gravity acting downwards} \\
\mathsf{D}_a = \mathsf{I} & \text{constitutive relations}\\ 
\mathsf{D}_b = \mathsf{I} \\ 
\mathsf{D}_c = \mathsf{I} \\ 
\mathsf{K} = \mathsf{I} \\ 
R(a,b) = ab \\
\mu = 1 \\
\rho(a,b,c) = a + \beta b + \gamma c & \text{$\gamma>\beta$ for destabilisation}\\
a_0(x,y)=\lim_{\epsilon\to0}\frac{1}{2}\left(1+\text{erf}\left(\frac{y-\sqrt{RaDa}}{\epsilon \sqrt{RaDa}}\right)\right)+\mathcal{N}(x,y) & \text{diffusive base state plus noise} \\
b_0>0 \\
c_0\geq0 \\
a_{\text{D}}(x,y=\sqrt{RaDa})=1 & \text{prescribed concentration on upper boundary} \\
a_{\text{N}}(x,y=0) & \text{no-flux on remaining boundary} \\
a_{\text{N}}(x=0,y) \\
a_{\text{N}}(x=\sqrt{RaDa}\mathcal{A},y) \\
b_{\text{N}}\vert_{\partial\Omega}=0 & \text{no-flux on entire boundary} \\
c_{\text{N}}\vert_{\partial\Omega}=0 & \text{no-flux on entire boundary} \\
\psi_{\text{D}}\vert_{\partial\Omega}=0 & \text{no-penetration on entire boundary}
\end{cases}
$$

```python
from lucifex.fdm import AB2, CN
from lucifex.sim import run
from lucifex.viz import plot_colormap, create_animation, save_figure, display_animation

from py.A07_darcy_abc_convection import darcy_abc_convection_rectangle


simulation = darcy_abc_convection_rectangle(
    Lx=3000,
    Ly=2000,
    Nx=64,
    Ny=64,
    cell='quadrilateral', 
    Pe=100.0,
    Lmbda=1.0, 
    D_adv=AB2,
    D_diff=CN,
)

n_stop = 1000
dt_init = 1e-6
n_init = 10
run(simulation, n_stop=n_stop, dt_init=dt_init, n_init=n_init)
```
