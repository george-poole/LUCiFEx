# Migration of buoyant CO<sub>2</sub>

##  Example: $d=2$ rectangle

constitutive relations
$$
\begin{align*}
\rho(c)&=c(1 - \beta c)^n \\
\mu&=1
\end{align*}
$$

initial condition

$$c_0(y) = \text{H}(\tfrac{1}{4}L_y-y)$$

no-flux boundary condition
$$\textbf{n}\cdot\nabla c\vert_{\partial\Omega}=0$$

no-penetration boundary condition
$$\textbf{n}\cdot\textbf{u}\vert_{\partial\Omega}=0\implies\psi\vert_{\partial\Omega}=0$$
