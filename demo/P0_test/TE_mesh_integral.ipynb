```python
from dolfinx.fem import FunctionSpace

from ufl.core.expr import Expr
from lucifex.mesh import rectangle_mesh, mesh_boundary
from lucifex.fem import SpatialFunction as Function
from lucifex.utils import integral
from lucifex.solver import integration


Lx = 2.0
Ly = 1.0
mesh = rectangle_mesh(Lx, Ly, 20, 10)
boundary = mesh_boundary(
    mesh, 
    {
        "left": lambda x: x[0],
        "right": lambda x: x[0] - Lx,
        "lower": lambda x: x[1],
        "upper": lambda x: x[1] - Ly,
    },
)


@integral
def multiply(a, b) -> Expr:
    return a * b

fs = FunctionSpace(mesh, ('P', 1))
u = Function(fs, lambda x: x[0], name='u')
w = Function(fs, lambda x: x[1], name='w')


uw_dx = multiply('ds', lambda x: x[1] - Ly, lambda x: x[1])(u, w)

uw_dx
```
