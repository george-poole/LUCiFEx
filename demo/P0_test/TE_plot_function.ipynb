# Visualization using the `lucifex.plt` module

```python
import numpy as np
from IPython.display import HTML
from dolfinx.fem import FunctionSpace
from lucifex.mesh import rectangle_mesh
from lucifex.fem import Function, cross_section
from lucifex.plt import plot_colormap, create_animation, plot_line, save_figure
```

## Colormaps

```python
mesh = rectangle_mesh(2.0, 1.0, 10, 10)
fs = FunctionSpace(mesh, ('P', 1))

f = Function(fs, lambda x: x[0]**2 - x[1]**2 + 10 * x[0] * x[1], name='f')
fig, ax = plot_colormap(f, title='$f(x,y)$')
```

## Cross-sections

```python
x_axis, fx, y_value = cross_section_grid(f, 'y', 0.5)
plot_line((x_axis, fx), x_label='$x$', y_label=f'${f.name}(y={y_value:.2f})$')

y_axis, fy, x_value = cross_section_grid(f, 'x', 0.5)
plot_line((y_axis, fy), x_label='$y$', y_label=f'${f.name}(x={x_value:.2f})$')
```

## Animation

```python
x_axis = np.linspace(0, 1, num=200)
times = np.linspace(0, 2, num=200)
fxt = lambda x, t: np.sin(2 * np.pi * (x - t))

line_series = [
    (x_axis, fxt(x_axis, t)) for t in times 
]
title_series = [f'$t={t:.2f}$' for t in times]

anim = create_animation(
    plot_line,
    x_label='$x$',
    y_label='$u(x,t)$',
)(line_series, title=title_series)
write(anim, f'{get_ipynb_file_name()}_animation', './figures')
```

```python
HTML(anim.to_html5_video())
```

```python
import numpy as np
from lucifex.plt import save_figure, plot_line_multifigure

x_axis = np.linspace(0, 2.0)
m_max_opts = (4, 5, 6)
lines = [
    [(x_axis, m * x_axis) for m in range(1, m_max + 1)] for m_max in m_max_opts
]
titles = [f'\n\\texttt{{m_max={m_max}}}' for m_max in m_max_opts]

fig, *_  = plot_line_multifigure(n_cols=1, cbars=True)(
    lines, 
    titles=titles, 
    cycs='jet',
    legend_labels=[(0, m * 3.14) for m in m_max_opts],
    legend_titles='$\ell$',
    x_label='$x$',
    y_label='$y$',
    x_lims=(0, 2),
    y_lims=(0, max(x_axis) * max(m_max_opts)),
)
```
