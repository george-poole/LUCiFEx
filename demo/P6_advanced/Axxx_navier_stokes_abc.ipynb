# ABC reaction in a Navier-Stokes fluid 

Langtangen, H. & Logg, A. (2016). *Solving PDEs in Python: The FEniCS Tutorial I*. $\S 3.5$

$$
\mathbb{S}_{\textbf{u},p,a,b,c}
\begin{cases}
\Sigma_a(a,b) = -kab + s_+(x,y) \\
\Sigma_b(a,b) = -kab + s_-(x,y) \\
\Sigma_c(a,b) = kab -kc \\
s_\pm(x,y) = \begin{cases}
0.1 & (x-x_0)^2 + (y\pm y_0)^2 < \epsilon^2 \\
0 & \text{otherwise}
\end{cases} \\
\end{cases}
$$
