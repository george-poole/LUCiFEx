# Meshes

```python
from lucifex.mesh import (
    interval_mesh, rectangle_mesh, annulus_mesh, ellipse_mesh, 
    circle_sector_mesh, annulus_sector_mesh, structured_annulus_mesh, rectangle_minus_ellipse_mesh,
)
from lucifex.viz import plot_mesh
from lucifex.mesh import as_quad_mesh

Lx = 2.0
Nx = 8
mesh = interval_mesh(Lx, Nx)
fig, ax = plot_mesh(mesh)
```

```python
Lx = 2.0
Ly = 1.0
Nx = 16
Ny = 16
cell_types = (
    'quadrilateral', 'right', 'left', 'left_right', 'right_left', 'crossed',
)
for cell in cell_types:
    mesh = rectangle_mesh(Lx, Ly, Nx, Ny, cell) 
    fig, ax = plot_mesh(mesh)
```

```python
h = 0.1
cell_types = ('triangle', 'quadrilateral')

radii = (2.0, 1.0)
for cell in cell_types:
    mesh = ellipse_mesh(h, cell)(radii)
    plot_mesh(mesh)

Rinner = 1.0
Router = 2.0
for cell in cell_types:
    mesh = annulus_mesh(h, cell)(Rinner, Router)
    plot_mesh(mesh)


Rinner = 1.0
Router = 2.0
Nr = 4
Ntheta_per_quadrant = 4
mesh = structured_annulus_mesh(cell='quadrilateral')(Rinner, Router, Nr, Ntheta_per_quadrant) 
plot_mesh(mesh)

# radius = 1.0
# angle = 45.0
# for cell in cell_types:
#     mesh = circle_sector_mesh(h, cell)(radius, angle)
#     plot_mesh(mesh)

Rinner = 1.0
Router = 2.0
angle = 45.0
for cell in cell_types:
    mesh = annulus_sector_mesh(h, cell)(Rinner, Router, angle)
    plot_mesh(mesh)

Lx = 2.0
Ly = 1.0
radii = (0.2, 0.1)
for cell in cell_types:
    mesh = rectangle_minus_ellipse_mesh(h, cell)(Lx, Ly, radii)
    plot_mesh(mesh)
```
