```python
import time
from lucifex.sim import run, as_grid_simulation
from lucifex.viz import plot_colormap
from py.U20_simulation import create_simulation


sim = create_simulation(store_delta=1)(
    Nx=150,
    Ny=150,
    dt=0.01,
    d=0.01,
)
run(sim, n_stop=200)


for use_mesh_map in (True, False):
    t_start = time.perf_counter()
    sim_grid = as_grid_simulation(sim, use_mesh_map=use_mesh_map)
    t_end = time.perf_counter()
    print(use_mesh_map, t_end - t_start)

```

```python
u = sim_grid['u']

plot_colormap(u.series[0])
plot_colormap(u.series[-1])
```
