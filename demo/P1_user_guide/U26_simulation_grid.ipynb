```python
import time
from lucifex.sim import run, as_grid_simulation
from py.U20_simulation import create_simulation


sim = create_simulation(store_delta=1)(
    Nx=150,
    Ny=150,
    dt=0.01,
    d=0.01,
)
run(sim, n_stop=200)

for use_mapping in (True, False):
    if use_mapping:
        use_dof_coordinates_opts = (False, True)
    else:
        use_dof_coordinates_opts = (False, )
    for use_dof_coordinates in use_dof_coordinates_opts:
        t_start = time.perf_counter()
        sim_grid = as_grid_simulation(
            sim, use_mapping=use_mapping, use_dof_coordinates=use_dof_coordinates,
        )
        t_end = time.perf_counter()
        print(use_mapping, use_dof_coordinates, t_end - t_start)
```
