# Running simulations in parallel

```python
import time
from lucifex.viz import plot_colormap
from lucifex.sim import parallel_run, as_grid_simulation, GridSimulation
from py.U27_simulation_parallel import create_simulation, performance_report 

STORE = 1
create_sim = create_simulation(store_delta=STORE)

N_STOP = 20
NX = 200
NY = 200

n_proc_opts = (3, None)
d_opts = (0.1, 1.0, 5.0, 10.0)
dt_opts = (0.01, 0.02)
results: dict[int | None, dict[float, GridSimulation]] = {}

for n_proc in n_proc_opts:
    t_start = time.perf_counter()
    simulations = parallel_run(
        create_sim, n_proc, N_STOP, serialize=as_grid_simulation, link=False,
    )(NX, NY)(d=d_opts, dt=dt_opts)
    t_stop = time.perf_counter()
    results[n_proc] = simulations
    t_exec = t_stop - t_start
    print(performance_report(t_exec, n_proc, N_STOP, STORE, NX, NY), '\n')
```
