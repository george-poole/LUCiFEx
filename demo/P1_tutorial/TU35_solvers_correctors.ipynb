## Dirichlet correction

```python
from lucifex.mesh import rectangle_mesh
from lucifex.fem import Function
from lucifex.solver import interpolation
from lucifex.pde.streamfunction_vorticity import velocity_from_streamfunction
from lucifex.utils.fenicsx_utils import dirichlet_corrector

Lx = 1.0
Ly = 1.0
mesh = rectangle_mesh(Lx, Ly, 32, 32)
psi = Function((mesh, 'P', 2), lambda x: x[0] * (Lx - x[0]) * x[1])

u_uncorrected = Function((mesh, 'P', 1, 2))
u_corrected = Function((mesh, 'P', 1, 2))

u_uncorrected_solver = interpolation(u_uncorrected, velocity_from_streamfunction)(psi)

corrector = dirichlet_corrector(
    u_corrected.function_space,
    (lambda x: x[0], 0.0, 0),
    (lambda x: x[0] - Lx, 0.0, 0),
)
u_corrected_solver = interpolation(u_corrected, velocity_from_streamfunction, corrector)(psi)

```
