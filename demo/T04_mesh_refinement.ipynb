```python
import numpy as np
from lucifex.mesh import refine, rectangle_mesh
from lucifex.utils import cell_sizes
from lucifex.viz import plot_mesh


Lx = 2.0
Ly = 1.0
cell = 'right'
name = 'Omega'
mesh = rectangle_mesh(Lx, Ly, 10, 10, cell, name)
hmin = np.min(cell_sizes(mesh, 'hmin'))
hmax = np.max(cell_sizes(mesh, 'hmax'))

plot_mesh(mesh, title='$\Omega$')
print(f'Before refining:\n hmin, hmax = {hmin:.2f}, {hmax:.2f}')
```

```python
marker = lambda x: x[1] > 0.75 * Ly
mesh_refined = refine('Omega_prime', mesh, marker)
hmin_refined = np.min(cell_sizes(mesh_refined, 'hmin'))
hmax_refined = np.max(cell_sizes(mesh_refined, 'hmax'))

plot_mesh(mesh_refined, title="$\Omega'$")
print(f'After refining:\n hmin, hmax = {hmin_refined:.2f}, {hmax_refined:.2f}')
```
