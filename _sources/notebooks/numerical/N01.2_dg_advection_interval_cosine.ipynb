## Advection of a cosine profile on an interval

c.f. Donea & Huerta, *Finite Element Methods for Flow Problems* $\S 3.11.1$


$$
\mathbb{S}
\begin{cases}
\Omega = [0, L_x] \\
\textbf{a}=a\,\textbf{e}_x \\
s = 0 \\
u_0(x)=\frac{1}{2}(1+\cos(\pi(x-x_0)/\sigma))~\text{if}~|x-x_0|\leq\sigma \\
u_0(x)=0~\text{if}~|x-x_0|>\sigma \\
u_{\text{I}} = 0 \\
\end{cases}
$$
